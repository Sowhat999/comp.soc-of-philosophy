```python
%matplotlib notebook  

from sklearn.decomposition import PCA
from sklearn.cross_decomposition import CCA

import networkx as nx
import pandas as pd
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
G = nx.read_gexf("datasets/influences.gexf")
```

```python
A = nx.to_numpy_matrix(G)
```

```python
pca = PCA()
Z = pca.fit(A).transform(A)
Z = zip(*Z)
```

```python
print len(pca.explained_variance_ratio_)

# i = 0
# s = 0
# while s < .99:
#   i = i + 1
#   s = np.sum(pca.explained_variance_ratio_[0:i])
  
#   print (i, s)

def plot_explained_vr(pca):
  (x, y) = zip(*list(enumerate(pca.explained_variance_ratio_)))
  
  y = [ np.sum(y[0:i]) for i, dummy in enumerate(y) ]
  
  fig = plt.figure()
  plt.plot(x, y)
  
  plt.show()

plot_explained_vr(pca)
```
